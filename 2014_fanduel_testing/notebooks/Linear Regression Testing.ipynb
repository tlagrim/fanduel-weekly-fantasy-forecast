###Convert a/h to 1/0
Here, away is 0, home is 1.

```python
wk1['homeaway'] = wk1['h/a']
wk1.loc[wk1['h/a'] == 'a', 'homeaway'] = 0
wk1.loc[wk1['h/a'] == 'h', 'homeaway'] = 1

wk1.plot(kind='scatter',x='homeaway',y='FD points')
```

There seems to be a small correlation between the two.

### putting both charts in one area:

```python
# visualize the relationship between the features and the response using scatterplots
fig, axs = plt.subplots(1, 2, sharey=True)
wk1.plot(kind='scatter', x='FD salary', y='FD points', ax=axs[0],figsize=(16, 8))
wk1.plot(kind='scatter', x='homeaway', y='FD points', ax=axs[1])
```

###This is a bit confusing, let's split it up this time
Also, why don't we make this graph look good.

```python
oppt = sns.stripplot(x='Oppt',y='FD points',data=wk1)
oppt.set_xticklabels(wk1['Oppt'],rotation=55)
```

Each data point above shows the amount of points a player scored on a given team. We can use this graph to visualize which teams allow the most points scored against them. On the other hand, let's look at points each player was able to score FOR their team.

```python
team = sns.stripplot(x='Team',y='FD points',data=wk1)
team.set_xticklabels(wk1['Team'],rotation=55)
```
