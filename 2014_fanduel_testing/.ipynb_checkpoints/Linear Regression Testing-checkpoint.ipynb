###Let's take away the spaces from FD points and FD salary

```python
wk1['FDpoints'] = wk1['FD points']
wk1['FDsalary'] = wk1['FD salary']
```

```python
import statsmodels.formula.api as smf

## create a fitted model in one line
lm = smf.ols(formula='FDpoints ~ FDsalary', data=wk1).fit()

## print the coefficients
lm.params
```

```python
# manually calculate the prediction
# how many points is the average player going to score
# if their salary is 6667?

-9.197744+0.002935*6667
```

#Plotting the Least Squares Line
Let's make predictions for the smallest and largest observed values of x, and then use the predicted values to plot the least squares line:

```python
## create a DataFrame with the minimum and maximum values of TV
X_new = pd.DataFrame({'FDsalary': [wk1.FDsalary.min(), wk1.FDsalary.max()]})
X_new.head()
```

```python
## make predictions for those x values and store them
preds = lm.predict(X_new)
preds
```

```python
# first, plot the observed data
wk1.plot(kind='scatter', x='FDsalary', y='FDpoints')

# then, plot the least squares line
plt.plot(X_new, preds, c='red', linewidth=2)
```

```python
# print the confidence intervals for the model coefficients
lm.conf_int()
```

```python
# print the p-values for the model coefficients
lm.pvalues
```

```python
# print the R-squared value for the model
lm.rsquared
```
